```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url=['https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&ref_=adv_prv','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt']
```

```python
#taking the urls into a page from the server
page=[]
for i in url:
    page.append(requests.get(i))
```

```python
#now using beautifulsoup
soup=[]
for i in page:
    soup.append(BeautifulSoup(i.text,'html.parser'))
```

```python
#determining the ID
id_text=[]
id=[]
for i in soup:
    id_text.append(i.find_all(class_='lister-item-index unbold text-primary'))
for item in id_text:
    for i in item:
        id.append(i.text)
```

```python
#determining the titles of the movies
title_text=[]
title=[]
for i in soup:
    title_text.append(i.find_all('h3',{'class':'lister-item-header'}))
for item in title_text:
    for i in item:
        title.append(i.find('a').text)
```

```python
len(title)
```

```python
#determining the Release years of the movies
Release_year_text=[]
Release_year=[]
for i in soup:
    Release_year_text.append(i.find_all(class_='lister-item-year text-muted unbold'))
for item in Release_year_text:
    for i in item:
        Release_year.append(i.text)
```

```python
len(Release_year)
```

```python
#determining the runtimes of the movies
runtime=[]
runtime_text=[]
for i in soup:
    runtime_text.append(i.find_all(class_='runtime'))
for item in runtime_text:
    for i in item:
        runtime.append(i.text)
```

```python
len(runtime)
```

```python
#determining the genres of the movies
genre=[]
genre_text=[]
for i in soup:
    genre_text.append(i.find_all(class_='genre'))
for item in genre_text:
    for i in item:
        genre.append(i.text)
for i,j in enumerate(genre):
    genre[i]=j.replace('\n','').strip()
```

```python
len(genre)
```

```python
#determining the ratings of the movies
rating_text=[]
rating=[]
for i in soup:
    rating_text.append(i.find_all('div',{'class':'inline-block ratings-imdb-rating'}))
for item in rating_text:
    for i in item:
        rating.append(i.text)
for i,j in enumerate(rating):
    rating[i]=j.replace('\n','').strip()
```

```python
len(rating)
```

```python
#determining the descriptions of the movies
description_text=[]
for i in soup:
    description_text.append(i.find_all('p',{'class':'text-muted'}))
description=[]
for item in description_text:
    for i in item:
        description.append(i.text)
d=[]
for i in range(len(description)):
    if(int(i)%2!=0):
        d.append(description[i])
for i,j in enumerate(d):
    d[i]=j.replace('\n','').strip()
```

```python
about=d
```

```python
#determining the directors of the movies
director_text=[]
for i in soup:
    director_text.append(i.find_all('p',{'class':''}))
director_star=[]
for item in director_text:
    for i in item:
        if 'Director' in i.text:
             director_star.append(i.text)
director=[]
for i in range(len(director_star)):
    director.append(director_star[i].split("Stars:")[0])
for i,j in enumerate(director):
    director[i]=j.replace('\n','').strip()
for i,j in enumerate(director):
    director[i]=j.replace('|','').strip()
for i,j in enumerate(director):
    director[i]=j.replace('Director:','').strip()
```

```python
len(director)
```

```python
#determininng the stars of the movies
stars=[]
for i in range(len(director_star)):
    stars.append(director_star[i].split("Stars:")[1])
for i,j in enumerate(stars):
    stars[i]=j.replace('\n','').strip()
```

```python
len(stars)
```

```python
#determining the votes of the movies
votes_gross_text=[]
for i in soup:
    votes_gross_text.append(i.find_all(class_='sort-num_votes-visible'))
votes_gross=[]
for item in votes_gross_text:
    for i in item:
        votes_gross.append(i.text)
votes=[]
for i in range(len(votes_gross)):
    votes.append(votes_gross[i].split("|")[0])
for i,j in enumerate(votes):
    votes[i]=j.replace('\n','').strip()
for i,j in enumerate(votes):
    votes[i]=j.replace('Votes:','').strip()
```

```python
len(votes)
```

```python
#determining the gross of the movies
gross=[]
for i in range(len(votes_gross)):
    if(len(votes_gross[i].split("|"))==2):
        gross.append(votes_gross[i].split("|")[1])
    else:
        gross.append(None)
for i,j in enumerate(gross):
    if(gross[i]!=None):
        gross[i]=j.replace('\n','').strip()
for i,j in enumerate(gross):
    if(gross[i]!=None):
        gross[i]=j.replace('Gross:','').strip()
```

```python
len(gross)
```

```python
final_array=[]
for i,ti,re,ru,g,ra,de,di,st,vo,gr in zip(id,title,Release_year,runtime,genre,rating,about,director,stars,votes,gross):
    final_array.append({'ID':i,'TITLE':ti,'RELEASE YEAR':re,'RUN TIME':ru,'GENRE':g,'RATING':ra,'DESCRIPTION':de,'DIRECTOR':di,'STARS':st,'VOTES':vo,'GROSS':gr})
```

```python
#converting to datafame
data=pd.DataFrame(final_array)
```

```python
data=data[['ID','TITLE','RELEASE YEAR','RUN TIME','GENRE','RATING','DESCRIPTION','DIRECTOR','STARS','VOTES','GROSS']]
```
